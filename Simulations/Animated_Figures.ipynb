```python
import git
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter

```

```python
fig = plt.figure()
l, = plt.plot([], [], 'k-')
plt.xlim(-5,5)
plt.ylim(-5,5)

#Function for what we are plotting - sine wave

def func(x):
  return np.sin(x)*3

xlist = np.linspace(-5,5,100)
ylist = func(xlist)

l.set_data(xlist, ylist)
plt.show()

#Create writer object that collects frames and puts them together into a video
metadata = dict(title='Movie', artist='coder')
writer = PillowWriter(fps=15, metadata=metadata)
#pillowwriter is good for GIFS, FFMPEG for mp4


#creating animations has multiple figures, each figure is generated with a list of data that progressively gets longer, stored in below lists
xlist = []
ylist = []

with writer.saving(fig, "sinWave.gif", 100):
  for xval in np.linspace(-5,5,100):
    xlist.append(xval)
    ylist.append(func(xval))

    l.set_data(xlist, ylist)

    writer.grab_frame() #grabs frame currently displayed in figure object
```

```python
from matplotlib import cm
np.random.seed(9384925)

metadata = dict(title='Movie', artist='artist')
writer = FFMpegWriter(fps=15, metadata=metadata)

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

plt.xlim(-5,5)
plt.ylim(-5,5)


def func3(x,y,r,t):
  return np.cos(r/2+t)*np.exp(-np.square(r)/50)

xvec = np.linspace(-10, 10, 1000)
yvec = np.linspace(-10, 10, 1000)

xlist, ylist = np.meshgrid(xvec, yvec)

rlist = np.sqrt(np.square(xlist) + np.square(ylist))

with writer.saving(fig, "Anim3d.mp4", 100):
  for tval in np.linspace(0,20,160):
    print(tval)
    zval = func3(xlist, ylist, rlist, tval)
    ax.set_zlim(-1,1)
    ax.plot_surface(xlist,ylist,zval, cmap=cm.viridis)

    writer.grab_frame()
    plt.cla()
```
